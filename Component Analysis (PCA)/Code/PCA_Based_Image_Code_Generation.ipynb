### 游섫릖 Code Explanation


This code performs Principal Component Analysis (PCA) on the grayscale image `bird` to reduce its dimensionality and then reconstructs the image using a reduced number of components (100 in this case).

*   `pca_50 = PCA(n_components = 100)`: Initializes a PCA object with 100 components.
*   `imagem_50 = pca_50.fit_transform(bird)`: Fits the PCA model to the image data and transforms the data into the reduced dimensionality space.
*   `imagem_nova = pca_50.inverse_transform(imagem_50)`: Reconstructs the image from the reduced dimensionality data back to the original space.
*   `plt.imshow(imagem_nova, cmap='gray_r')`: Displays the reconstructed image in grayscale.
*   `plt.axis('off')`: Hides the axes for a cleaner image display.
*   `plt.show()`: Shows the plot.

```python
bird = plt.imread('passaro.jpg') #load image

R,G,B = bird[:,:,0] , bird [:,:,1], bird[:,:,2]
bird = 0.2989 * R + 0.5870 * G + 0.1140 * B


# Show Image
plt.imshow(bird, cmap = 'gray')
plt.axis('off') # Hide the Axes
plt.show()

bird.shape # Show Image Size
```

```python
pca2 = PCA(n_components=256)
pca2.fit(bird)

plt.grid()
plt.plot(np.cumsum(pca2.explained_variance_ratio_* 100))
plt.ylabel('Explained Variance')
plt.ylabel('Explained Variance')
```

Note that with 100 components, you can explain almost 100% of the image variance.

PCA with More Components




```python
pca_more = PCA(n_components = 250)
imagem_more = pca_more.fit_transform(bird)
imagem_nova_more = pca_more.inverse_transform(imagem_more)
plt.imshow(imagem_nova_more, cmap='gray_r')
plt.axis('off')
plt.show()
```

PCA with Fewer Components

```python
pca_20 = PCA(n_components = 20)
imagem_20 = pca_20.fit_transform(bird)
imagem_nova_20 = pca_20.inverse_transform(imagem_20)
plt.imshow(imagem_nova_20, cmap='gray_r')
plt.axis('off')
plt.show()
```

```python
# Calculate the difference between the original and reconstructed images
error_20 = np.abs(bird - imagem_nova_20)
error_100 = np.abs(bird - imagem_nova) # Note: imagem_nova might be from 250 components if overwritten
error_250 = np.abs(bird - imagem_nova_more)

# Display the error images side-by-side
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Display error for 20 components
axes[0].imshow(error_20, cmap='gray')
axes[0].set_title('Error (20 Components)')
axes[0].axis('off')

# Display error for 100 components
axes[1].imshow(error_100, cmap='gray')
axes[1].set_title('Error (100 Components)') # Title might be inaccurate if imagem_nova was overwritten
axes[1].axis('off')

# Display error for 250 components
axes[2].imshow(error_250, cmap='gray')
axes[2].set_title('Error (250 Components)')
axes[2].axis('off')

plt.tight_layout()
plt.show()
```
